# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt

# Query to get ConvertedComp data
query_converted_comp = """
SELECT ConvertedComp
FROM master
"""

# Fetch data into a DataFrame
df_converted_comp = pd.read_sql_query(query_converted_comp, conn)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df_converted_comp['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# Query to get Age data
query_age = """
SELECT Age
FROM master
"""

# Fetch data into a DataFrame
df_age = pd.read_sql_query(query_age, conn)

# Plot box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df_age['Age'], vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Query to get Age and WorkWeekHrs data
query_age_workweek = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Fetch data into a DataFrame
df_age_workweek = pd.read_sql_query(query_age_workweek, conn)

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_age_workweek['Age'], df_age_workweek['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Query to get WorkWeekHrs, CodeRevHrs, and Age data
query_bubble = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# Fetch data into a DataFrame
df_bubble = pd.read_sql_query(query_bubble, conn)

# Plot bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df_bubble['WorkWeekHrs'], df_bubble['CodeRevHrs'], s=df_bubble['Age'], alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs (Age as Bubble Size)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Query to get data of 'WorkWeekHrs' and 'CodeRevHrs' for the age group 30 to 35
query_stacked = """
SELECT Age, 
       AVG(WorkWeekHrs) as AvgWorkWeekHrs, 
       AVG(CodeRevHrs) as AvgCodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
"""

# Fetch data into a DataFrame
df_stacked = pd.read_sql_query(query_stacked, conn)

# Plot the stacked chart
plt.figure(figsize=(10, 6))
plt.bar(df_stacked['Age'], df_stacked['AvgWorkWeekHrs'], color='b', label='Avg WorkWeekHrs')
plt.bar(df_stacked['Age'], df_stacked['AvgCodeRevHrs'], color='r', label='Avg CodeRevHrs', bottom=df_stacked['AvgWorkWeekHrs'])
plt.title('Stacked Chart of Avg WorkWeekHrs and Avg CodeRevHrs (Age Group: 30-35)')
plt.xlabel('Age')
plt.ylabel('Average Hours')
plt.legend()
plt.grid(True)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Query to get data of 'ConvertedComp' for the age group 45 to 60
query_median_converted_comp = """
SELECT Age, AVG(ConvertedComp) as MedianConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
GROUP BY Age
"""

# Fetch data into a DataFrame
df_median_converted_comp = pd.read_sql_query(query_median_converted_comp, conn)

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_median_converted_comp['Age'], df_median_converted_comp['MedianConvertedComp'], marker='o', color='g', linestyle='-')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Query to get data of the 'MainBranch' column
query_main_branch = """
SELECT MainBranch, COUNT(*) as count
FROM master
GROUP BY MainBranch
"""

# Fetch data into a DataFrame
df_main_branch = pd.read_sql_query(query_main_branch, conn)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(df_main_branch['MainBranch'], df_main_branch['count'], color='skyblue')
plt.title('Horizontal Bar Chart of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(axis='x')
plt.show()
```
